```python
import io, os, numpy as np, pandas as pd
import matplotlib.pyplot as plt, seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
from sklearn.feature_selection import mutual_info_classif
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import joblib

assert os.path.exists(csv_path), "CSV not found — fix csv_path or upload again."
data = pd.read_csv(csv_path)
# downcast
for c in data.select_dtypes('float64').columns: data[c] = data[c].astype('float32')
for c in data.select_dtypes('int64').columns:   data[c] = data[c].astype('int32')
if 'CLASS_LABEL' in data.columns and 'labels' not in data.columns:
    data.rename(columns={'CLASS_LABEL': 'labels'}, inplace=True)

data.head()
```

```python
ax = data['labels'].value_counts().sort_index().plot(kind='bar', rot=0)
ax.set_title("Label Distribution (0=Legit, 1=Phishing)"); ax.set_xlabel("Label"); ax.set_ylabel("Count"); plt.show()

def corr_heatmap(df, start, end):
    temp = df.iloc[:, start:end].copy()
    if 'id' in temp.columns: temp = temp.drop(columns=['id'])
    temp['labels'] = df['labels']
    corr = temp.corr(method='spearman')
    plt.figure(figsize=(10,8)); sns.heatmap(corr, annot=False, cmap='viridis')
    plt.title(f"Spearman Correlation (cols {start}–{end})"); plt.show()

corr_heatmap(data, 0, 10); corr_heatmap(data, 10, 20); corr_heatmap(data, 20, 30)
corr_heatmap(data, 30, 40); corr_heatmap(data, 40, len(data.columns))
```

```python
X = data.drop(columns=['labels'], errors='ignore').copy()
if 'id' in X.columns: X = X.drop(columns=['id'])
y = data['labels']

discrete_features = np.array([pd.api.types.is_integer_dtype(dt) for dt in X.dtypes])
mi = mutual_info_classif(X, y, discrete_features=discrete_features, random_state=42)
mi_scores = pd.Series(mi, index=X.columns, name="MI Scores").sort_values(ascending=False)
mi_scores.head(15)
plt.figure(figsize=(8, max(4, len(mi_scores)*0.25))); mi_scores.sort_values().plot(kind='barh')
plt.title("Mutual Information Scores"); plt.tight_layout(); plt.show()
```

```python
def evaluate_topn(clf, features_sorted, data, topn_range=range(20, 51)):
    rows = []
    for n in topn_range:
        feats = list(features_sorted[:n])
        X, y = data[feats], data['labels']
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        rows.append({
            "num_features": n,
            "precision": precision_score(y_test, y_pred),
            "recall":    recall_score(y_test, y_pred),
            "f1":        f1_score(y_test, y_pred),
            "accuracy":  accuracy_score(y_test, y_pred),
        })
    return pd.DataFrame(rows)
```

```python
log_df = evaluate_topn(LogisticRegression(max_iter=10000), mi_scores.index, data, topn_range=range(20, 51))
display(log_df)
plt.plot(log_df['num_features'], log_df['precision'], label='Precision')
plt.plot(log_df['num_features'], log_df['recall'], label='Recall')
plt.plot(log_df['num_features'], log_df['f1'], label='F1')
plt.plot(log_df['num_features'], log_df['accuracy'], label='Accuracy')
plt.title("Logistic Regression vs. Top-N features"); plt.xlabel("Top-N"); plt.ylabel("Score"); plt.legend(); plt.show()
log_df.iloc[log_df['accuracy'].idxmax()]
```

```python
rf_df = evaluate_topn(RandomForestClassifier(n_estimators=300, n_jobs=-1, random_state=42), mi_scores.index, data, topn_range=range(20, 51))
display(rf_df)
plt.plot(rf_df['num_features'], rf_df['precision'], label='Precision')
plt.plot(rf_df['num_features'], rf_df['recall'], label='Recall')
plt.plot(rf_df['num_features'], rf_df['f1'], label='F1')
plt.plot(rf_df['num_features'], rf_df['accuracy'], label='Accuracy')
plt.title("Random Forest vs. Top-N features"); plt.xlabel("Top-N"); plt.ylabel("Score"); plt.legend(); plt.show()
best_rf_row = rf_df.iloc[rf_df['accuracy'].idxmax()]
best_rf_row
```

```python
best_n = int(best_rf_row['num_features'])
final_features = list(mi_scores.index[:best_n])
X, y = data[final_features], data['labels']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)
final_rf = RandomForestClassifier(n_estimators=300, n_jobs=-1, random_state=42)
final_rf.fit(X_train, y_train)
y_pred = final_rf.predict(X_test)
print(classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred)
fig, ax = plt.subplots(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", xticklabels=["Legit (0)", "Phishing (1)"], yticklabels=["Legit (0)", "Phishing (1)"], ax=ax)
ax.set_title("Confusion Matrix - Final RandomForest"); ax.set_xlabel("Predicted"); ax.set_ylabel("True"); plt.tight_layout(); plt.show()

import os, joblib
os.makedirs("outputs", exist_ok=True)
fig.savefig("outputs/confusion_matrix.png", dpi=150)
with open("outputs/classification_report.txt", "w") as f: f.write(classification_report(y_test, y_pred))
joblib.dump(final_rf, "outputs/rf_phishing_model.pkl")
with open("outputs/selected_features.txt", "w") as f: f.write("\n".join(final_features))
print("Saved artifacts in ./outputs")
```
